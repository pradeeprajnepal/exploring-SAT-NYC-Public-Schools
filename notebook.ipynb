```python

import pandas as pd


schools = pd.read_csv("schools.csv")


schools.head()
```

```python
total_score = 800
best_math_schools = schools[schools["average_math"] > total_score*0.8][["school_name","average_math"]].sort_values("average_math", ascending= False)
best_math_schools
```

```python
import numpy as np
#with df method
schools["total_SAT"]= schools[["average_math","average_reading","average_writing"]].sum(axis=1)
top_10_schools= schools.sort_values("total_SAT", ascending=False)[["school_name","total_SAT"]].head(10)
top_10_schools

```

```python

boroughs = schools.groupby('borough')["total_SAT"].agg(["count", "mean", "std"]).round(2)
print(boroughs.head())


largest_std_dev_filter = boroughs[boroughs["std"]== boroughs['std'].max()]


largest_std_dev= largest_std_dev_filter.rename(columns={'count':'num_schools', 'mean':'average_SAT', 'std':'std_SAT'})


largest_std_dev


```
